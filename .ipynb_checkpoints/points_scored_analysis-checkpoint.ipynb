```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read in our data
df_list = []
for i in range(15,21):
    df = pd.read_csv(f'NBA_PBP_20{i}-{i+1}.csv')
    df_list.append(df)
    
df = pd.concat(df_list)

#Look at what datatypes we're working with
df.dtypes
```

```python
df['Date'].head()
```

```python
#Create some useful features
#Only regular time, no OT
df = df.loc[(df['Quarter']<=4)]
#Convert quarter # to quarters left
df['qtrs_left'] = df['Quarter']*-1 + 4
#Total time left, 720 is how many seconds in a quarter
df['total_sec_left'] = df['qtrs_left']*720 + df['SecLeft']
df['total_min_left'] = df['total_sec_left']/60
#Score difference
df['score_diff'] = df['AwayScore'] - df['HomeScore']
df['abs_score_diff'] = np.abs(df['score_diff'])
#Make sure the data is sorted chronologically
df.sort_values(by=['URL','total_sec_left'], ascending=False, inplace=True)
```

```python
#Grab final play of the game
final_play = df.groupby(by=['URL']).tail(1)
#User final play to determine final result of each game
final_play['result'] = final_play['score_diff']
final_play['abs_result'] = final_play['abs_score_diff']
#Merge game final result back with original df
result_df = pd.merge(df, final_play[['URL','result','abs_result']], on='URL', how='left')
```

```python
#We only want close games in regular season
close_games = result_df.loc[(result_df['abs_result']<=5) & (result_df['GameType']=='regular')]

#Get the total points in the game
close_games['total_points'] = close_games['AwayScore'] + close_games['HomeScore']
#Get the differences between points, accounting for borders of new games
close_games['points_scored'] = close_games.groupby(['URL'])['total_points'].transform(lambda x: x.diff()) 
```

```python
#Bin our time remaining to every 30 seconds
cut_bins = np.linspace(0, 2880, 97)
labels = cut_bins.astype(int)
labels = labels[0:-1]

close_games['binned_sec_left'] = pd.cut(close_games['total_sec_left'], bins=cut_bins, labels=labels, include_lowest=True)

#Get summed points scored per 30 seconds on a game_level
ps_games = close_games.groupby(by=['URL','binned_sec_left'])['points_scored'].sum().to_frame().reset_index()
```

```python
close_games.loc[(close_games['binned_sec_left']!=close_games['binned_sec_left'])]
```

```python
ps_games.head(3)
```

```python
#Now bin and average across games
bin_avg = ps_games.groupby(by=['binned_sec_left'])['points_scored'].mean().to_frame().reset_index()
```

```python
%%time
#Plotting
plt.style.use('default')
plt.style.use('seaborn-talk')
# plt.style.use('dark_background')
plt.title('Regular season NBA games Oct. 2015 - Jan. 2021, end-of-regulation point differential <=5 (n=1,785)', fontsize=10)
plt.suptitle('Scoring volume increases at the end of games', fontsize=24)
plt.scatter(bin_avg['binned_sec_left'], bin_avg['points_scored'], c='#EEE1C6', label='Avg points scored in 30 second bins')

sns.regplot(x='binned_sec_left', 
            y='points_scored', 
            scatter=False, 
            data=ps_games, 
            order=12, 
            label='Smoothed polynomial',
            line_kws = {'color': '#00471B', 'alpha': 1, 'ls':'-'})

plt.axvline(720, ls='--', c='k', alpha=0.1)
plt.axvline(1440, ls='--', c='k', alpha=0.1)
plt.axvline(2160, ls='--', c='k', alpha=0.1, label='End of Quarter')
plt.axvline(0, ls='--', c='k', alpha=0.1)

plt.ylabel("Points Scored Per 30 Seconds\n", fontsize=18)
plt.xlabel('\nGame Seconds Remaining', fontsize=18)



plt.xlim(2880,-50)
plt.legend()

txt= "Data: Kaggle via Pro Basketball Reference                                           Chart:@EthanCDouglas"
plt.figtext(0.11, -0.06, txt, wrap=True, horizontalalignment='left', fontsize=12)
```
